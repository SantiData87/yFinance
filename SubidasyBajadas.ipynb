```python
import yfinance as yf
import pandas as pd
from datetime import date

TICKER = 'ggal.ba'
# Fecha de hoy
today = date.today()

# Descargar datos de una acción (puedes cambiar el ticker y el rango de fechas)
ticker = TICKER  # Cambia esto por el ticker de tu interés
data = yf.download(ticker, start="2010-01-01", end=today)

# Calcular los retornos porcentuales diarios
data['Retorno'] = data['Close'].pct_change() * 100  # Retorno diario en porcentaje

# Filtrar los días que subió más de un 5%
dias_subida_5 = data[data['Retorno'] > 5]

# Filtrar los días que bajó más de un 5%
dias_bajada_5 = data[data['Retorno'] < -5]
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Descargar datos de una acción (puedes cambiar el ticker y el rango de fechas)
ticker =  TICKER  # Cambia esto por el ticker de tu interés
data = yf.download(ticker, start="2023-01-01", end=today)

# Calcular los retornos porcentuales diarios
data['Retorno'] = data['Close'].pct_change() * 100  # Retorno diario en porcentaje

# Filtrar los días que subió más de un 5%
dias_subida_5 = data[data['Retorno'] > 5]

# Filtrar los días que bajó más de un 5%
dias_bajada_5 = data[data['Retorno'] < -5]

# Graficar el precio de cierre con los días resaltados
plt.figure(figsize=(10, 6))

# Gráfico de línea para los precios de cierre
plt.plot(data.index, data['Close'], label='Precio de Cierre', color='blue')

# Marcar los días que subieron más del 5%
plt.scatter(dias_subida_5.index, dias_subida_5['Close'], color='green', label='Subidas > 5%', marker='^', s=100)

# Marcar los días que bajaron más del 5%
plt.scatter(dias_bajada_5.index, dias_bajada_5['Close'], color='red', label='Bajadas > 5%', marker='v', s=100)

# Añadir etiquetas y título
plt.title(f'{ticker}: Días con Subidas y Bajadas mayores al 5%')
plt.xlabel('Fecha')
plt.ylabel('Precio de Cierre (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import yfinance as yf
import pandas as pd
import mplfinance as mpf

# Descargar datos de una acción (puedes cambiar el ticker y el rango de fechas)
ticker = TICKER  # Cambia esto por el ticker de tu interés
data = yf.download(ticker, start="2020-01-01", end=today)

# Asegurarse de trabajar sobre una copia del DataFrame para evitar el warning
data = data.copy()

# Calcular los retornos porcentuales diarios
data['Retorno'] = data['Close'].pct_change() * 100  # Retorno diario en porcentaje

# Filtrar los días que subió más de un 5% y los que bajó más de un 5%
dias_subida_5 = data[data['Retorno'] > 5].copy()  # Crear copia
dias_bajada_5 = data[data['Retorno'] < -5].copy()  # Crear copia

# Calcular los precios promedio de cierre para las subidas y bajadas
media_subida = dias_subida_5['Close'].mean()
media_bajada = dias_bajada_5['Close'].mean()

# Crear columnas adicionales necesarias para el gráfico de velas
data['Date'] = data.index
data['Date'] = pd.to_datetime(data['Date'])
data = data[['Open', 'High', 'Low', 'Close', 'Volume']]

# Graficar con mplfinance y resaltar los promedios de las subidas y bajadas mayores al 5%
mpf.plot(data, type='candle', style='charles', volume=True,
         title=f'{ticker} Precio de Cierre con Movimientos de >5%',
         datetime_format='%Y-%m-%d',
         hlines=dict(hlines=[media_subida, media_bajada],
                     colors=['g', 'r'], linestyle='--'))
```
