```python
import yfinance as yf
import pandas as pd
```

```python
# prompt: crear una unica tabla obteniendo en las columnas los precios Close de los siguiente tickers 'MSFT','GOOG','AAPL','AMZN', 'META', 'TSLA', 'NVDA'



# Create a list of tickers
tickers = ['MSFT','GOOG','AAPL','AMZN', 'META', 'TSLA', 'NVDA']

# Download historical data for each ticker
historical_data = {}
for ticker in tickers:
    historical_data[ticker] = yf.download(ticker, period="2y", interval="1d")['Close']

# Create a DataFrame with the historical data
df = pd.DataFrame(historical_data)

# Print the DataFrame
print(df)
```

```python
# prompt: mostrar el dataframe anteror en forma tabla para interactuar
from google.colab import data_table

# Display the DataFrame as a table
data_table.DataTable(df)
```

```python
# prompt: agrupar la tabla por ticker creando una sola tabla

import pandas as pd

# Create a list of dataframes
dataframes = []
for ticker in tickers:
  dataframes.append(yf.Ticker(ticker).history(period='YTD'))

# Concatenate the dataframes into a single dataframe
df = pd.concat(dataframes, axis=1)

# Group the dataframe by the 'Symbol' column
df = df.groupby('Symbol')

# Print the grouped dataframe
print(df)
```

```python
# prompt: generar el grafico anterior pero todos los tickers en el mimso grafico

import yfinance as yf
import matplotlib.pyplot as plt

tickers = ['MSFT','GOOG','AAPL','AMZN', 'META', 'TSLA', 'NVDA']
plt.figure(figsize=(16, 10))

for i, ticker in enumerate(tickers):
  current_ticker =yf.Ticker(ticker)
  plt.plot(current_ticker.history(period='365d')['Close'], label=ticker)

plt.xlabel("Date")
plt.ylabel("Close Price")
plt.title("Precios 7Magnificas 1 año")
plt.legend()
plt.show()
```

```python
# prompt: obtener el promedio de lo anterior

import yfinance as yf
import matplotlib.pyplot as plt

# Calculate the average closing price for each ticker over the last 10 years.
tickers = ['MSFT','GOOG','AAPL','AMZN', 'META', 'TSLA', 'NVDA']
average_closing_prices = []
for ticker in tickers:
  current_ticker = yf.Ticker(ticker)
  average_closing_prices.append(current_ticker.history(period='10y')['Close'].mean())

# Print the average closing prices.
print(average_closing_prices)

# Calculate the overall average closing price.
overall_average_closing_price = sum(average_closing_prices) / len(average_closing_prices)

# Print the overall average closing price.
print(overall_average_closing_price)
```
