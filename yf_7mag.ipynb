# Grafico 7mag

```python
#Todos los tickers en un mismo grafico

import yfinance as yf
#Importar librerias de Visualizacion
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

#Estilo de gafico y tamaño
mpl.style.use('_mpl-gallery-nogrid')  #Definir el Estilo del grafico
plt.figure(figsize=(18, 8)) # Definir el tamaño de la figura

tickers = ['MSFT','GOOG','AAPL','AMZN', 'META', 'TSLA', 'NVDA']
plt.figure(figsize=(16, 10))

for i, ticker in enumerate(tickers):
  current_ticker =yf.Ticker(ticker)
  plt.plot(current_ticker.history(period='365d')['Close'], label=ticker)

plt.xlabel("Date")
plt.ylabel("Close Price")
plt.title("Precios 7Magnificas 1 año")
plt.legend()
plt.show()

plt.savefig('7mag.png')
```

# Graficos apilados 

```python
#Cada uno de los tickers en un propio grafico, uno arriba del otro

import yfinance as yf
tickers = ['MSFT','GOOG','AAPL','AMZN', 'META', 'TSLA', 'NVDA']

import matplotlib.pyplot as plt
for i, ticker in enumerate(tickers):
  current_ticker = yf.Ticker(ticker)
  plt.subplot(len(tickers),1,i+1)
  current_ticker.history(period='365d')['Close'].plot(figsize= (16,60), title ='Precios 7Magnificas 1 año '+ticker)
 
plt.show()
```
