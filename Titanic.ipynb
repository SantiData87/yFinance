## def

```python
def pay_if_reach(x):
  if x['Pclass'] == 1:
    return x['Fare'] * 1.2
  else:
    return x['Fare']
```

```python
df['Fare_for_reach'] = df.apply(pay_if_reach, axis=1)
```

```python
df.head()
```

## lambda
Son funciones anonimas porque no tienene nombre.
Sirve cuando aplico una funcion ua sola vez y no voy a necesitarla despues
La funcion solo vive en la linea de codigo utlizada y no puede ser "llamada" de nuevo

```python
df['Fare_for_reach_2'] = df.apply(lambda x: x['Fare'] * 1.2 if x['Pclass'] == 1 else x['Fare'] , axis=1)
```

# Merge
Es lo mismo que JOIN en SQL

```python
# Creao un DataFrame
df_class = pd.DataFrame({'Pclass': [1,2,3], 'Clase' : ['Alta', 'Media', 'Baja']})
df_class
```

```python
# Left significa que tra TODOS los registros de 'df' y solo los de 'df_class? en donde coincide la columna 'Pclass'
df_merge = df.merge(df_class, how='left', on='Pclass')
df_merge
```
